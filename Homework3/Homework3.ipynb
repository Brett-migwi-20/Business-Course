```python
import matplotlib.pyplot as plt

# Data for the bubble chart provided as a string
data_string = """
College	Top 10% HS	Acceptance Rate	Expenditures/Student
A	86	21%	$26,708
B	68	54%	$17,822
C	60	35%	$17,739
D	96	37%	$23,563
E	76	23%	$26,033
"""

data_string = data_string.replace(",","").replace("\t",",").replace("%","").replace("$","")

# Split the data string into lines
lines = data_string.strip().split('\n')

# Extract the header and data
header = lines[0].split(',')
data_lines = lines[1:]

# Initialize lists to store the data
colleges = []
top_10_percent_hs = []
acceptance_rate = []
expenditures_per_student = []

# Iterate through each line and append the data to the lists
for line in data_lines:
    values = line.split(',')
    colleges.append(values[0])
    top_10_percent_hs.append(int(values[1]))
    acceptance_rate.append(int(values[2]))
    expenditures_per_student.append(int(values[3]))

# Increase the size of the bubbles by adjusting the scaling factor
plt.figure(figsize=(10, 6))
plt.scatter(top_10_percent_hs, acceptance_rate, 
            s=[expenditure / 2 for expenditure in expenditures_per_student],  # Increased bubble size scaling
            alpha=0.6, edgecolors="w", linewidth=2)

# Add labels to each bubble
for i, college in enumerate(colleges):
    plt.text(top_10_percent_hs[i], acceptance_rate[i], college, fontsize=12, ha='center', va='center')

# Label axes
plt.xlabel('Top 10% HS (%)')
plt.ylabel('Acceptance Rate (%)')
plt.title('Bubble Chart of College Data with Increased Bubble Size')
plt.grid(True)

# Show the plot
plt.show()

```
