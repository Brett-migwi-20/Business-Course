## QUESTION 1

Create clustered column and stacked column charts for the​ pre-MBA and​ post-MBA salary data in the accompanying dataset. Discuss which type of chart you feel is better to use to explain the information.


```python
import pandas as pd
import matplotlib.pyplot as plt
import io

# Data in string format
data = """
Individual	Age	Years of Work Experience	Pre-MBA Salary	Post-MBA Salary
1	30	More than 6	86,000	105,000
2	34	More than 6	84,000	105,000
3	28	4 to 6	70,000	105,000
4	35	More than 6	95,000	125,000
5	23	1 to 3	55,000	70,000
6	28	4 to 6	51,000	70,000
7	39	More than 6	70,000	120,000
8	26	4 to 6	55,000	75,000
9	30	More than 6	76,000	110,000
10	24	1 to 3	50,000	80,000
11	23	1 to 3	43,500	85,000
12	28	4 to 6	70,000	95,000
13	28	4 to 6	55,000	70,000
14	27	4 to 6	61,000	95,000
15	26	4 to 6	67,000	80,000
"""

data = data.replace(",","").replace("\t",",")

# Read data from string into a pandas DataFrame
df = pd.read_csv(io.StringIO(data))

# Create Clustered Column Chart
plt.figure(figsize=(10, 6))
plt.bar(df['Individual'] - 0.2, df['Pre-MBA Salary'], width=0.4, label='Pre-MBA Salary', color='blue')
plt.bar(df['Individual'] + 0.2, df['Post-MBA Salary'], width=0.4, label='Post-MBA Salary', color='red')
plt.xlabel('Individual')
plt.ylabel('Salary')
plt.title('Clustered Column Chart: Pre-MBA vs Post-MBA Salary')
plt.xticks(df['Individual'])
plt.legend()
plt.show()

# Create Stacked Column Chart
plt.figure(figsize=(10, 6))
plt.bar(df['Individual'], df['Pre-MBA Salary'], label='Pre-MBA Salary', color='blue')
plt.bar(df['Individual'], df['Post-MBA Salary'] , bottom=df['Pre-MBA Salary'], label='Post-MBA Salary', color='red')
plt.xlabel('Individual')
plt.ylabel('Salary')
plt.title('Stacked Column Chart: Pre-MBA vs Post-MBA Salary')
plt.xticks(df['Individual'])
plt.legend()
plt.show()
```

### Part C

Cluster column is better because it is better for comparing the differences

```python
import matplotlib.pyplot as plt

# Data for the bubble chart provided as a string
data_string = """
College	Top 10% HS	Acceptance Rate	Expenditures/Student
A	86	23%	$26,408
B	70	53%	$17,906
C	57	34%	$18,003
D	95	38%	$23,843
E	78	24%	$25,467
"""

data_string = data_string.replace(",","").replace("\t",",").replace("%","").replace("$","")

# Split the data string into lines
lines = data_string.strip().split('\n')

# Extract the header and data
header = lines[0].split(',')
data_lines = lines[1:]

# Initialize lists to store the data
colleges = []
top_10_percent_hs = []
acceptance_rate = []
expenditures_per_student = []

# Iterate through each line and append the data to the lists
for line in data_lines:
    values = line.split(',')
    colleges.append(values[0])
    top_10_percent_hs.append(int(values[1]))
    acceptance_rate.append(int(values[2]))
    expenditures_per_student.append(int(values[3]))

# Increase the size of the bubbles by adjusting the scaling factor
plt.figure(figsize=(10, 6))
plt.scatter(top_10_percent_hs, acceptance_rate, 
            s=[expenditure / 2 for expenditure in expenditures_per_student],  # Increased bubble size scaling
            alpha=0.6, edgecolors="w", linewidth=2)

# Add labels to each bubble
for i, college in enumerate(colleges):
    plt.text(top_10_percent_hs[i], acceptance_rate[i], college, fontsize=12, ha='center', va='center')

# Label axes
plt.xlabel('Top 10% HS (%)')
plt.ylabel('Acceptance Rate (%)')
plt.title('Bubble Chart of College Data with Increased Bubble Size')
plt.grid(True)

# Show the plot
plt.show()

```
