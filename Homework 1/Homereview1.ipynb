Suppose that a manufacturer can produce a part for ​$12.00 with a fixed cost of ​$6,000. ​Alternately, the manufacturer could contract with a supplier in Asia to purchase the part at a cost of ​$15.00​, which includes transportation.

a. If the anticipated production volume is 1,300 ​units, compute the total cost of manufacturing and the total cost of outsourcing.

b. What is the best​ decision?

```python
# Given values
fixed_cost = 2000  # Fixed cost of manufacturing in dollars
variable_cost_manufacturing = 9.00  # Cost per unit for manufacturing in dollars
outsourcing_cost_per_unit = 10.00  # Cost per unit for outsourcing in dollars
production_volume = 1400  # Anticipated production volume in units

# Calculate total cost of manufacturing
total_cost_manufacturing = fixed_cost + (variable_cost_manufacturing * production_volume)

# Calculate total cost of outsourcing
total_cost_outsourcing = outsourcing_cost_per_unit * production_volume

print(f"The total cost of manufacturing is ${total_cost_manufacturing}")

print(f"The total cost of outsourcing is ​$ ${total_cost_outsourcing}")


if(total_cost_manufacturing > total_cost_outsourcing):
    print("The best decision is the outsourcing, because it has the lower total cost.")
else:
    print("The best decision is the manufacturing, because it has the lower total cost.")

    
```

# QUESTION 10

The accompanying table shows all of the items available to a hiker that will aid their hike out of the wilderness. Containers of food and water provide​ energy, shelter will protect the hiker from the​ elements, and defense will protect the hiker from wild animals. Each item has a weight and survival points. The hiker must take one item from each of the four categories​ (food, water,​ shelter, defense).​ Unfortunately, the​ hiker's backpack has a maximum capacity of 33 lbs. The​ hiker's chance for survival is calculated by adding all of the survival points together from the chosen items. What is the maximum chance for survival the hiker can​ achieve?

```python
# Define the available items with their weights and survival points
items = {
    'food': [(5, 10), (8, 20), (12, 25)],        # (weight, survival points) for food items
    'water': [(3, 10), (5, 20), (8, 25)],        # (weight, survival points) for water items
    'shelter': [(5, 5), (8, 15), (12, 20)],      # (weight, survival points) for shelter items
    'defense': [(1, 5), (2, 15), (3, 20)]        # (weight, survival points) for defense items
}

# Define the maximum capacity of the backpack
max_capacity = 25

# Initialize variable to store the maximum survival points
max_survival_points = 0

# Iterate over all combinations of items
for food in items['food']:
    for water in items['water']:
        for shelter in items['shelter']:
            for defense in items['defense']:
                # Calculate total weight and survival points for the current combination
                total_weight = food[0] + water[0] + shelter[0] + defense[0]
                total_survival_points = food[1] + water[1] + shelter[1] + defense[1]
                
                # Check if the total weight is within the backpack capacity
                if total_weight <= max_capacity:
                    # Update the maximum survival points if the current combination is better
                    max_survival_points = max(max_survival_points, total_survival_points)

max_survival_points
```
