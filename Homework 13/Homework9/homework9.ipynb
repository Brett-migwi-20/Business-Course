```python
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt

# Data as a string
data_string = """
Date	Price
5/4/2009	2.05
5/11/2009	2.22
5/18/2009	2.28
5/25/2009	2.41
6/1/2009	2.50
6/8/2009	2.60
6/15/2009	2.64
6/22/2009	2.65
6/29/2009	2.59
7/6/2009	2.56
7/13/2009	2.48
7/20/2009	2.41
"""

data_string = data_string.replace("\t",",")

# Reading the data into a pandas DataFrame
data = pd.read_csv(StringIO(data_string), parse_dates=['Date'])

# Creating a three-period moving average and rounding to 2 decimal places
data['3-Period Moving Average'] = data['Price'].rolling(window=3).mean().round(2)

# Displaying the resulting DataFrame
print(data)

# Plotting the original prices and the 3-period moving average for visualization
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Price'], marker='o', linestyle='-', label='Original Price')
plt.plot(data['Date'], data['3-Period Moving Average'], marker='o', linestyle='--', color='red', label='3-Period Moving Average')
plt.title('Gasoline Prices with 3-Period Moving Average')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate the Mean Absolute Deviation (MAD)
data['Absolute Deviation'] = abs(data['Price'] - data['3-Period Moving Average'])

# Compute the MAD, ignoring NaN values
mad = data['Absolute Deviation'].mean()

# Round MAD to three decimal places
mad_rounded = round(mad, 3)

print(f"MAD = {mad_rounded}")
```

```python
# Calculate the Mean Squared Error (MSE)
data['Squared Error'] = (data['Price'] - data['3-Period Moving Average']) ** 2

# Compute the MSE, ignoring NaN values
mse = data['Squared Error'].mean()

# Round MSE to three decimal places
mse_rounded = round(mse, 3)


print(f"MAPE = {mse_rounded}")
```

```python
# Calculate the Mean Absolute Percentage Error (MAPE)
data['Percentage Error'] = abs((data['Price'] - data['3-Period Moving Average']) / data['Price']) * 100

# Compute the MAPE, ignoring NaN values
mape = data['Percentage Error'].mean()

# Round MAPE to three decimal places
mape_rounded = round(mape, 3)

print(f"MAPE = {mape_rounded}")
```

```python
import pandas as pd
import numpy as np
from io import StringIO

# Data as a string
data_string = """
Year	Coal Production (thousands of tons)
1	434,330
2	420,425
3	439,039
4	477,195
5	504,182
6	526,954
7	546,818
8	564,886
9	556,710
10	570,978
"""

data_string = data_string.replace(",","").replace("\t",",")

# Read the data into a pandas DataFrame
data = pd.read_csv(StringIO(data_string))

# Double Exponential Smoothing Function
def double_exponential_smoothing(series, alpha, beta):
    n = len(series)
    level = [series[0]]  # Initialize a_1 = A_1
    trend = [series[1] - series[0]]  # Initialize b_1 = A_2 - A_1
    forecast = [series[0]]  # Initialize F_1 = A_1
    
    for t in range(1, n):
        # Calculate the new level and trend based on the given formulas
        level_new = alpha * series[t] + (1 - alpha) * (level[-1] + trend[-1])
        trend_new = beta * (level_new - level[-1]) + (1 - beta) * trend[-1]
        forecast.append(level_new + trend_new)
        level.append(level_new)
        trend.append(trend_new)
    
    return np.array(forecast)

# Calculate MAD for each pair of alpha and beta
def calculate_mad(actual, forecast):
    mad = np.mean(np.abs(actual - forecast)).round(2)
    return mad

# Define alpha and beta pairs
alpha_beta_pairs = [(0.2, 0.9), (0.4, 0.3), (0.9, 0.7)]

# Calculate MAD for each alpha and beta pair dynamically
mad_values = []
for alpha, beta in alpha_beta_pairs:
    forecast = double_exponential_smoothing(data['Coal Production (thousands of tons)'], alpha, beta)
    mad = calculate_mad(data['Coal Production (thousands of tons)'].values, forecast)
    mad_values.append(mad)

# Displaying the MAD values
for i, (alpha, beta) in enumerate(alpha_beta_pairs):
    print(f"Alpha: {alpha}, Beta: {beta}, MAD: {mad_values[i]}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error

# Creating the dataset
data = {
    'Date': ['1/7/2005', '1/14/2005', '1/21/2005', '1/28/2005', '2/4/2005', '2/11/2005', '2/18/2005', '2/25/2005', '3/4/2005', '3/11/2005', '3/18/2005', '3/25/2005', '4/1/2005', '4/8/2005', '4/15/2005', '4/22/2005', '4/29/2005', '5/6/2005', '5/13/2005', '5/20/2005', '5/27/2005', '6/3/2005', '6/10/2005', '6/17/2005', '6/24/2005', '7/1/2005', '7/8/2005', '7/15/2005', '7/22/2005', '7/29/2005', '7/1/2005', '7/8/2005', '7/15/2005', '7/22/2005'],
    'Rate': [5.21, 5.19, 5.15, 5.14, 5.14, 5.10, 5.14, 5.22, 5.33, 5.38, 5.47, 5.56, 5.58, 5.48, 5.46, 5.36, 5.33, 5.31, 5.33, 5.27, 5.21, 5.20, 5.14, 5.22, 5.16, 5.12, 5.20, 5.25, 5.32, 5.34, 5.12, 5.20, 5.25, 5.32]
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Visualize the data to check for trend and seasonality
plt.figure(figsize=(10, 6))
plt.plot(df['Rate'], label='Mortgage Rate')
plt.title('Mortgage Rate over Time')
plt.xlabel('Date')
plt.ylabel('Rate')
plt.legend()
plt.show()

# Split the data into train and test sets
train = df['Rate'][:len(df)-6]
test = df['Rate'][len(df)-6:]

# Simple Exponential Smoothing
ses_model = ExponentialSmoothing(train, trend=None, seasonal=None, smoothing_level=0.3, optimized=False)
ses_fit = ses_model.fit()
ses_forecast = ses_fit.forecast(steps=len(test))

# Double Exponential Smoothing
des_model = ExponentialSmoothing(train, trend='add', seasonal=None, smoothing_level=0.3, smoothing_slope=0.2, optimized=False)
des_fit = des_model.fit()
des_forecast = des_fit.forecast(steps=len(test))

# Triple Exponential Smoothing (Holt-Winters)
hw_model = ExponentialSmoothing(train, trend='add', seasonal='add', smoothing_level=0.3, smoothing_slope=0.2, smoothing_seasonal=0.8, optimized=False)
hw_fit = hw_model.fit()
hw_forecast = hw_fit.forecast(steps=len(test))

# ARIMA Model (p=1, d=1, q=1)
arima_model = ARIMA(train, order=(1, 1, 1))
arima_fit = arima_model.fit()
arima_forecast = arima_fit.forecast(steps=len(test))

# Calculate Mean Absolute Error (MAD)
def calculate_mad(actual, forecast):
    return np.mean(np.abs(actual - forecast))

# MAD for each model
ses_mad = calculate_mad(test, ses_forecast)
des_mad = calculate_mad(test, des_forecast)
hw_mad = calculate_mad(test, hw_forecast)
arima_mad = calculate_mad(test, arima_forecast)

print(f"SIMPLE EXPONENTIAL SMOOTHING MAD: {ses_mad:.4f}")
print(f"DOUBLE EXPONENTIAL SMOOTHING MAD: {des_mad:.4f}")
print(f"HOLT-WINTERS ADDITIVE SEASONALITY MAD: {hw_mad:.4f}")
print(f"ARIMA MAD: {arima_mad:.4f}")

# Select the model with the minimum MAD
models = {'SES': ses_mad, 'DES': des_mad, 'Holt-Winters': hw_mad, 'ARIMA': arima_mad}
best_model = min(models, key=models.get)
print(f"Best model based on MAD: {best_model}")
```
