```python
"""
price_demand_revenue.py

Solves the airline price-demand problem and prints the answers (well-labeled, in order).
Also displays:
 - Price vs Demand plot with the two given points and fitted linear demand
 - Total Revenue vs Price plot

Given:
  P1 = 600  -> D1 = 440
  P2 = 400  -> D2 = 1160
"""

import numpy as np
import matplotlib.pyplot as plt

# Given data
P1, D1 = 600.0, 440
P2, D2 = 400.0, 1160.0

# -------------------------
# Part (a): Plot points and choose graph
# -------------------------
# Compute linear demand function D(P) = m*P + b
m = (D2 - D1) / (P2 - P1)            # slope
b = D1 - m * P1                      # intercept

# Make nice strings for equations
demand_eq = f"D(P) = {m:.3f} P + {b:.3f}"
# Simplify decimals to typical form (we expect m = -3.4, b = 2575)
demand_eq_simple = f"D(P) = {m:.1f} P + {b:.0f}"

# Determine which multiple-choice graph matches: we want points at (400,1215) and (600,535)
# That corresponds to option C in the problem statement
mc_choice = "C"

# Print Part a answers
print("PART (a) — Plot points and choose correct graph")
print("  Given points (Price, Demand):")
print(f"    Point 1: (P = {P1:.0f}, D = {D1:.0f})")
print(f"    Point 2: (P = {P2:.0f}, D = {D2:.0f})")
print()
print("  Correct multiple-choice graph: Option", mc_choice)
print()
print("  Linear demand function (fitted through the two points):")
print("  Exact (3 decimals):", demand_eq)
print("  Simplified:", demand_eq_simple)
print("\n" + "-"*70 + "\n")

# -------------------------
# Part (b): Total revenue as a function of price
# TR(P) = P * D(P) = P * (m P + b) = m P^2 + b P
# Note m is negative; write TR(P) = m*P^2 + b*P
# -------------------------
# Compute symbolic coefficients for TR
m_tr = m
b_tr = b
# TR(P) = m*P^2 + b*P; it's conventional to show leading coefficient negative
tr_eq = f"TR(P) = ({m_tr:.4f}) P^2 + ({b_tr:.3f}) P"
tr_eq_simpler = f"TR(P) = {m_tr:.1f} P^2 + {b_tr:.0f} P"
# But because m is negative, better to display as TR(P) = -3.4 P^2 + 2575 P
tr_eq_nice = f"TR(P) = {m_tr:.1f} P^2 + {b_tr:.0f} P"

print("PART (b) — Total revenue as a function of price")
print("  Using D(P) from part (a):")
print("  Demand:      ", demand_eq_simple)
print("  Total revenue:", tr_eq_nice)
print("    (Expanded exact form: ", tr_eq, ")")
print("\n" + "-"*70 + "\n")

# -------------------------
# Plots
# -------------------------
# Price range and demand line
P_vals = np.linspace(0, 1400, 400)
D_vals = m * P_vals + b

# Plot Price (x) vs Demand (y)
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(P_vals, D_vals, label=f"Demand line: {demand_eq_simple}", color='tab:blue')
ax.scatter([P1, P2], [D1, D2], color='red', zorder=5)
ax.annotate(f"({int(P1)},{int(D1)})", (P1, D1), textcoords="offset points", xytext=(5,-12))
ax.annotate(f"({int(P2)},{int(D2)})", (P2, D2), textcoords="offset points", xytext=(5,6))
ax.set_xlim(0, 1400)
ax.set_ylim(0, 1400)
ax.set_xticks(np.arange(0, 1401, 200))
ax.set_yticks(np.arange(0, 1401, 200))
ax.set_xlabel("Price (P) — dollars", fontsize=12)
ax.set_ylabel("Demand (D) — passengers per day", fontsize=12)
ax.set_title("Part (a): Price vs Demand (P horizontal, D vertical)\nPoints: (600,535) and (400,1215)", fontsize=13)
ax.grid(which='major', linestyle='--', alpha=0.5)
ax.legend()
plt.show()

# Plot Total Revenue TR(P)
TR_vals = P_vals * (m * P_vals + b)
fig2, ax2 = plt.subplots(figsize=(8,6))
ax2.plot(P_vals, TR_vals, color='tab:green', label=f"TR(P) = {m:.1f} P^2 + {b:.0f} P")
ax2.set_xlim(0, 1400)
# set y-limits to something reasonable based on computed TR
ax2.set_ylim(min(TR_vals)-1000, max(TR_vals)+1000)
ax2.set_xticks(np.arange(0, 1401, 200))
ax2.set_xlabel("Price (P) — dollars", fontsize=12)
ax2.set_ylabel("Total Revenue (TR) — dollars per day", fontsize=12)
ax2.set_title("Part (b): Total Revenue vs Price", fontsize=13)
ax2.grid(which='major', linestyle='--', alpha=0.5)
ax2.legend()
plt.show()

# -------------------------
# Useful numeric checks (optional prints)
# -------------------------
# Print numeric forms cleanly
print("Numeric checks / final forms:")
print(f"  Demand function: D(P) = {m:.3f}*P + {b:.3f}    (=> D(P) = {m:.1f}P + {b:.0f})")
print(f"  Total revenue: TR(P) = {m:.4f}*P^2 + {b:.3f}*P    (=> TR(P) = {m:.1f}P^2 + {b:.0f}P)")
print()
print("You can use these functions to compute demand or revenue at any price.")
print("Example: at P = $500:")
P_test = 500.0
D_test = m * P_test + b
TR_test = P_test * D_test
print(f"  D(500) = {D_test:.1f} passengers/day")
print(f"  TR(500) = ${TR_test:,.2f} per day")
```

```python
import math

def calculate_distance(x1, y1, x2, y2):
    """Calculate the straight-line distance between two points (x1, y1) and (x2, y2)."""
    return math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

def total_weighted_distance(restaurant_location, neighborhoods):
    """Calculate the total weighted distance from the restaurant to all neighborhoods."""
    x_r, y_r = restaurant_location
    total_distance = 0
    for (x, y, population) in neighborhoods:
        distance = calculate_distance(x_r, y_r, x, y)
        total_distance += distance * population
    return total_distance

# Data for neighborhoods: (X-coordinate, Y-coordinate, Population)

neighborhoods = [
    (3, 12, 1),  # Liberty
    (7, 5, 2),   # Jefferson
    (2, 1, 8)    # Adams
]

# Calculate the population-weighted average coordinates
total_population = sum(pop for _, _, pop in neighborhoods)
x_avg = sum(x * pop for x, _, pop in neighborhoods) / total_population
y_avg = sum(y * pop for _, y, pop in neighborhoods) / total_population

# Test locations
test_locations = [
    (0, 0),
    (round(x_avg), round(y_avg)),
    (2, 1)
]

# Evaluate each location to find the one with the minimum weighted distance
best_location = None
min_distance = float('inf')

for location in test_locations:
    distance = total_weighted_distance(location, neighborhoods)
    if distance < min_distance:
        min_distance = distance
        best_location = location

# Calculate the total weighted distance at the best location
total_distance_at_best_location = total_weighted_distance(best_location, neighborhoods)

# Round the result to two decimal places
total_distance_at_best_location_rounded = round(total_distance_at_best_location, 2)
total_distance_at_best_location_rounded
```

```python
def calculate_revenue(price_per_room, number_of_rooms_rented):
    """Calculate total revenue based on price per room and rooms rented."""
    return price_per_room * number_of_rooms_rented

def calculate_profit(total_revenue, operating_costs):
    """Calculate profit based on total revenue and fixed operating costs."""
    return total_revenue - operating_costs

def evaluate_profit_range(start, end, price_per_room, operating_costs):
    """Calculate total revenue and profit for each number of rooms rented in the given range."""
    results = []
    for rooms in range(start, end + 1):
        revenue = calculate_revenue(price_per_room, rooms)
        profit = calculate_profit(revenue, operating_costs)
        results.append({
            'Rooms': rooms,
            'Total_Revenue': revenue,
            'Profit': round(profit, 2)  # round to 2 decimal places
        })
    return results

# Given values
price_per_room = 945
operating_costs = 21000
rooms_range = (32, 50)

# Evaluate profit for each number of rooms rented in the specified range
profits_for_range = evaluate_profit_range(*rooms_range, price_per_room, operating_costs)

# Display the results
for row in profits_for_range:
    print(f"Rooms Rented: {row['Rooms']},  Profit: {row['Profit']:,.2f}")

# Specific result for 41 rooms
rooms_target = 41

target_row = next((r for r in profits_for_range if r['Rooms'] == rooms_target), None)
if target_row:
    print("\nSpecific case:")
    print(f"  If {rooms_target} rooms are rented:")
    print(f"    Total Revenue = ${target_row['Total_Revenue']:,.2f}")
    print(f"    Profit        = {target_row['Profit']:,.2f}")
else:
    rooms_min, rooms_max = rooms_range
    print(f"\nNo data for {rooms_target} rooms (range is {rooms_min}-{rooms_max}).")
```

```python
import pandas as pd

# Define the data for each scenario
scenarios = {
    'Scenario': ['Likely', 'Optimistic', 'Pessimistic'],
    'Expected Crowd': [2800, 4800, 1600],
    'Concessions Expenditure': [25, 30, 10],
    'Fixed Cost': [15000, 11000, 26000]
}

# Create a DataFrame to hold the scenarios
df = pd.DataFrame(scenarios)

# Constants
ticket_price = 20  # Ticket price per person
profit_percentage = 0.85  # Profit percentage

# Calculate revenues and profit for each scenario
df['Ticket Revenue'] = df['Expected Crowd'] * ticket_price
df['Concession Revenue'] = df['Expected Crowd'] * df['Concessions Expenditure']
df['Profit'] = profit_percentage * (df['Ticket Revenue'] + df['Concession Revenue']) - df['Fixed Cost']

# Display the profitability for each scenario
likely_profit = df.loc[df['Scenario'] == 'Likely', 'Profit'].values[0]
likely_profit_rounded = round(likely_profit, 2)

print(likely_profit_rounded)


# Retrieve and round the profitability for the Optimistic scenario
optimistic_profit = df.loc[df['Scenario'] == 'Optimistic', 'Profit'].values[0]
optimistic_profit_rounded = round(optimistic_profit, 2)

print(optimistic_profit_rounded)


# Retrieve and round the profitability for the Pessimistic scenario
pessimistic_profit = df.loc[df['Scenario'] == 'Pessimistic', 'Profit'].values[0]
pessimistic_profit_rounded = round(pessimistic_profit, 2)

print(pessimistic_profit_rounded)
```

```python
# Install PuLP if needed
# !pip install pulp

import pulp

# Create LP problem: maximize total expected return
model = pulp.LpProblem("Portfolio_Optimization", pulp.LpMaximize)

# Decision variables (amount invested in each asset)
Life      = pulp.LpVariable("Life_Insurance", lowBound=1250, upBound=2500)
Bond      = pulp.LpVariable("Bond_Mutual_Funds", lowBound=15000)        # No upper bound
Stock     = pulp.LpVariable("Stock_Mutual_Funds", lowBound=7500)       # No upper bound
Savings   = pulp.LpVariable("Savings_Account", lowBound=0)             # No lower/upper bound

# Total available investment
TOTAL_FUNDS = 100000

# Returns (as decimals)
r_life    = 0.055
r_bond    = 0.08
r_stock   = 0.11
r_savings = 0.045

# Risk factors
risk_life    = -0.6
risk_bond    = 1.9
risk_stock   = 2.2
risk_savings = -0.3

# ----------------------------
# Objective: maximize total return
# ----------------------------
model += (
    r_life * Life +
    r_bond * Bond +
    r_stock * Stock +
    r_savings * Savings
)

# ----------------------------
# Constraints
# ----------------------------

# Total funds must equal 50,000
model += Life + Bond + Stock + Savings == TOTAL_FUNDS

# Weighted risk per dollar ≤ 1.0
model += (
    (risk_life * Life +
     risk_bond * Bond +
     risk_stock * Stock +
     risk_savings * Savings) / TOTAL_FUNDS
    <= 1.0
)

# Solve
model.solve()

# Retrieve optimal results
life_val    = Life.value()
bond_val    = Bond.value()
stock_val   = Stock.value()
savings_val = Savings.value()

total_return = (
    r_life * life_val +
    r_bond * bond_val +
    r_stock * stock_val +
    r_savings * savings_val
)

# Display results
print(f"Optimal Solution:")
print(f"  Life Insurance:       {life_val:,.2f}")
print(f"  Bond Mutual Funds:    {bond_val:,.2f}")
print(f"  Stock Mutual Funds:   {stock_val:,.2f}")
print(f"  Savings Account:      {savings_val:,.2f}")
print(f"\nTotal Expected Return:  {total_return:,.2f}")
```
