```python
import pandas as pd
import numpy as np
from scipy import stats

# Data input as a string
data_string = """
Employee\tYears_of_Service\tGraduate_Degree
1	21	N
2	6	N
3	8	N
4	20	N
5	8	N
6	5	N
7	26	Y
8	9	Y
9	7	Y
10	8	Y
11	10	N
12	16	Y
13	8	Y
14	26	N
15	8	Y
16	9	N
17	28	N
18	19	Y
19	27	N
20	19	Y
"""

data_string = data_string.replace("\t",",")


# Convert the data string into a DataFrame
from io import StringIO

data = pd.read_csv(StringIO(data_string))

# a. Find a 99% confidence interval for the mean years of service
years_of_service = data['Years_of_Service']

# Calculate the sample mean and standard deviation
mean_years = years_of_service.mean()
std_years = years_of_service.std(ddof=1)
sample_size = len(years_of_service)

# Calculate the standard error
standard_error_years = std_years / np.sqrt(sample_size)

# Calculate the t-score for 99% confidence level
confidence_level = 0.95
t_score = stats.t.ppf((1 + confidence_level) / 2, df=sample_size-1)

# Calculate the margin of error
margin_of_error_years = t_score * standard_error_years

# Calculate the confidence interval
lower_bound_years = mean_years - margin_of_error_years
upper_bound_years = mean_years + margin_of_error_years

# Round the results to two decimal places
lower_bound_years_rounded = round(lower_bound_years, 2)
upper_bound_years_rounded = round(upper_bound_years, 2)

# b. Find a 99% confidence interval for the proportion of employees who have a graduate degree
graduate_degree = data['Graduate_Degree'].apply(lambda x: 1 if x == 'Y' else 0)

# Calculate the sample proportion
p_hat = graduate_degree.mean()

# Calculate the standard error for proportion
standard_error_proportion = np.sqrt((p_hat * (1 - p_hat)) / sample_size)

# Calculate the z-score for 99% confidence level (since proportion uses z-distribution)
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the margin of error for the proportion
margin_of_error_proportion = z_score * standard_error_proportion

# Calculate the confidence interval for the proportion
lower_bound_proportion = p_hat - margin_of_error_proportion
upper_bound_proportion = p_hat + margin_of_error_proportion

# Round the results to two decimal places
lower_bound_proportion_rounded = round(lower_bound_proportion, 2)
upper_bound_proportion_rounded = round(upper_bound_proportion, 2)

# Output both confidence intervals
print(f"99% Confidence Interval for Years of Service: [{lower_bound_years_rounded}, {upper_bound_years_rounded}]")

print("\nA. We are 99% confident that the population mean years of service will be between the confidence interval limits.")
print(f"\n99% Confidence Interval for Proportion of Graduate Degrees: [{lower_bound_proportion_rounded}, {upper_bound_proportion_rounded}]")
print("\nA. We are 99% confident that the population proportion of employees who have a graduate degree will be between the confidence interval limits.")
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the custom data and samples
data_custom = pd.read_excel('Quiz/custom.xlsx')
data_50 = pd.read_excel('Quiz/data-50.xlsx')
data_250 = pd.read_excel('Quiz/data-250.xlsx')

# -----------------------------------------------------------
# Part 1: Point estimates for the mean & standard deviation
# -----------------------------------------------------------
mean_custom = round(data_custom.mean().values[0], 3)
std_dev_custom = round(data_custom.std(ddof=1).values[0], 3)

# -----------------------------------------------------------
# Part 3: Means & std dev for samples of size 50
# -----------------------------------------------------------
sample_means_50 = []
sample_stds_50 = []

for i in range(1, 6):
    sample_means_50.append(round(data_50[f'Sample {i}'].mean(), 3))
    sample_stds_50.append(round(data_50[f'Sample {i}'].std(ddof=1), 3))

# -----------------------------------------------------------
# Part 4: Means & std dev for samples of size 250
# -----------------------------------------------------------
sample_means_250 = []
sample_stds_250 = []

for i in range(1, 6):
    sample_means_250.append(round(data_250[f'Sample {i}'].mean(), 3))
    sample_stds_250.append(round(data_250[f'Sample {i}'].std(ddof=1), 3))

# -----------------------------------------------------------
# Part 5: Empirical rule (mean ± 3*SD)
# -----------------------------------------------------------
mean_of_50 = np.mean(sample_means_50)
std_of_50 = np.mean(sample_stds_50)
lower_bound_50 = round(mean_of_50 - 3 * std_of_50, 3)
upper_bound_50 = round(mean_of_50 + 3 * std_of_50, 3)

mean_of_250 = np.mean(sample_means_250)
std_of_250 = np.mean(sample_stds_250)
lower_bound_250 = round(mean_of_250 - 3 * std_of_250, 3)
upper_bound_250 = round(mean_of_250 + 3 * std_of_250, 3)

# -----------------------------------------------------------
# FORMATTED OUTPUT
# -----------------------------------------------------------

print("\n" + "="*65)
print("                POINT ESTIMATES (CUSTOM DATA)")
print("="*65)
print(f"{'Mean:':30} {mean_custom}")
print(f"{'Standard Deviation:':30} {std_dev_custom}")

print("\n" + "="*65)
print("                SAMPLE SIZE 50 STATISTICS")
print("="*65)
for i in range(5):
    print(f"Sample {i+1} Mean: {sample_means_50[i]:8}   |   Std Dev: {sample_stds_50[i]}")

print("\n" + "="*65)
print("                SAMPLE SIZE 250 STATISTICS")
print("="*65)
for i in range(5):
    print(f"Sample {i+1} Mean: {sample_means_250[i]:8}   |   Std Dev: {sample_stds_250[i]}")

print("\n" + "="*65)
print("                EMPIRICAL RULE (±3 SD)")
print("="*65)
print(f"Sample Size 50 Interval :   [{lower_bound_50}, {upper_bound_50}]")
print(f"Sample Size 250 Interval:   [{lower_bound_250}, {upper_bound_250}]")
print("="*65 + "\n")
```
